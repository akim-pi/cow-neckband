```python
from skimage.exposure import rescale_intensity
from skimage.feature import hog

def convertHOG(classN):
    #hog_image = []
    hog_feature = []
    for i in range(0, len(classN)):
        fd, hog_image = hog(classN[i], 
                            orientations = 8, 
                            pixels_per_cell = (16, 16), 
                            visualize = True)
        #hog_image.append(hog_image)
        hog_feature.append(fd)
    return hog_feature
```

```python
class3 = [skimage.io.imread(file, as_gray=True) for file in data_dir.glob("class6/*.bmp")]
hog_feat= convertHOG(class3)
hogf = np.array(hog_feat)
np.savetxt("hog6.txt", hogf, delimiter =", ")
files.download("hog6.txt")
```
