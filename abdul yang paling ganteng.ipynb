```python
crop_img = img[200:700, 300:900]
plt.imshow(crop_img, cmap="gray")
```

```python
def increase_brightness(i, value):
  lim = 255 - value
  i[i > lim] = 255
  i[i <= lim] += value

  return i
```

```python
bright = increase_brightness(crop_img, 150)
plt.imshow(bright, cmap="gray")
```
